```python
import pandas as pd
import re
import requests
```

```python
dico = {
        'Année' : [],
        'Commune': [],
        'Code INSEE' : [],
        'Département' : [],
        'Numéro département' : [],
        'Volume' : [],
        'Usage' : [],
        'Mode obtention' : [],
        'Longitude' : [],
        'Latitude' : [],
        'Statut' : [],
        'Qualification' : [],
        'Prélèvement écrasant' : [],
        'Ouvrage' : [],
        'Geometrie' : [],
        'url_fiche' : []

}
```

```python
for year in range(2008, 2023):
    url = "https://hubeau.eaufrance.fr/api/v1/prelevements/chroniques?annee=" +str(year) + "&code_departement=&format=json&size=10000"
    docu = requests.get(url).json()
    pages = re.search(r'page=(\d+)', docu['last'])
    nb_pages = int(pages.group(1))
    for page in range(1, nb_pages+1):
        url2 = "https://hubeau.eaufrance.fr/api/v1/prelevements/chroniques?annee=" +str(year) + "&code_departement=&format=json&page=" + str(page) + "&size=10000"
        docu2 = requests.get(url2).json()
        for prel in docu2['data']:
            dico['Année'].append(prel['annee'])
            dico['Commune'].append(prel['nom_commune'])
            dico['Code INSEE'].append(prel['code_commune_insee'])
            dico['Département'].append(prel['libelle_departement'])
            dico['Numéro département'].append(prel['code_departement'])
            dico['Volume'].append(prel['volume'])
            dico['Usage']. append(prel['libelle_usage'])
            dico['Mode obtention'].append(prel['libelle_mode_obtention_volume'])
            dico['Longitude'].append(prel['longitude'])
            dico['Latitude'].append(prel['latitude'])
            dico['Statut'].append(prel['libelle_statut_volume'])
            dico['Qualification'].append(prel['libelle_qualification_volume'])
            dico['Prélèvement écrasant'].append(prel['prelevement_ecrasant'])
            dico['Ouvrage'].append(prel['nom_ouvrage'])
            dico['Geometrie'].append(prel['geometry'])
            dico['url_fiche'].append(prel['uri_ouvrage'])



```

```python
df= pd.DataFrame(dico)
```

```python
url_departement = "https://geo.api.gouv.fr/regions/11/departements"
response_departement = requests.get(url_departement) 
data_departement = response_departement.json()
list_dep = []
for dep in data_departement:
        list_dep.append(dep['code'])
list_dep
```

```python
prelevement_idf = prelevement[prelevement['Numéro département'].isin(list_dep)]
prelevement_idf 
```

```python
prelevement_idf.to_csv("export/a_ignorer/prelevement_idf.csv", sep=";", index=False)
```

```python
for dep in list_dep : 
    prelevement_idf_dep = prelevement_idf[prelevement_idf['Numéro département'] == dep]
    prelevement_idf_dep.to_csv(f"export/prelevement_idf{dep}.csv", sep=";", index=False)
```

```python
for dep in list_dep : 
    prelevement_idf_dep_last = prelevement_idf[(prelevement_idf['Numéro département'] == dep) & (prelevement_idf['Année'] == prelevement_idf['Année'].max())]
    prelevement_idf_dep_last.to_csv(f"export/prelevement_idf_last{dep}.csv", sep=";", index=False)
```
