# Les données sont inutilisables car trop de trous, il va falloir extraire directement de l'API hubeau

```python
url = "https://hubeau.eaufrance.fr/api/v1/prelevements/chroniques?annee=2008&code_departement=&format=json&size=50"
docu = requests.get(url).json()
docu['data'][0]
```

```python
# Dictionnaire vide qui servira de base au futur df
dico = {
        'Année' : [],
        'Commune': [],
        'Code INSEE' : [],
        'Département' : [],
        'Numéro département' : [],
        'Volume' : [],
        'Usage' : [],
        'Mode obtention' : [],
        'Longitude' : [],
        'Latitude' : [],
        'Statut' : [],
        'Qualification' : [],
        'Prélèvement écrasant' : [],
        'Ouvrage' : []

}
```

```python
# Extraction des données de l'API dans le dictionnaire puis mise en place du DataFrame pour nettoyage
for year in range(2008, 2023):
    url = "https://hubeau.eaufrance.fr/api/v1/prelevements/chroniques?annee=" +str(year) + "&code_departement=&format=json&size=50"
    docu = requests.get(url).json()
    pages = re.search(r'page=(\d+)', docu['last'])
    nb_pages = int(pages.group(1))
    for page in range(1, nb_pages+1):
        url2 = "https://hubeau.eaufrance.fr/api/v1/prelevements/chroniques?annee=" +str(year) + "&code_departement=&format=json&page=" + str(page) + "&size=50"
        docu2 = requests.get(url2).json()
        for prel in docu2['data']:
            dico['Année'].append(prel['annee'])
            dico['Commune'].append(prel['nom_commune'])
            dico['Code INSEE'].append(prel['code_commune_insee'])        
            dico['Département'].append(prel['libelle_departement'])
            dico['Numéro département'].append(prel['code_departement'])
            dico['Volume'].append(prel['volume'])
            dico['Usage']. append(prel['libelle_usage'])
            dico['Mode obtention'].append(prel['libelle_mode_obtention_volume'])
            dico['Longitude'].append(prel['longitude'])
            dico['Latitude'].append(prel['latitude'])
            dico['Statut'].append(prel['libelle_statut_volume'])
            dico['Qualification'].append(prel['libelle_qualification_volume'])
            dico['Prélèvement écrasant'].append(prel['prelevement_ecrasant'])
            dico['Ouvrage'].append(prel['nom_ouvrage'])

df = pd.DataFrame(dico)
```

```python
df_base = pd.read_csv("C:/Users/Antoine/Documents/AquaVision/donnees/brutes/eau_potable/communes_AEP_2023.csv", encoding="Windows-1252", sep=";", low_memory=False)
# Filtrage du df pour récupérer les colonnes que l'on veut
df_base = df_base[["Code INSEE de la commune adhérente", "Identifiant SISPEA de la collectivité de l'entité de gestion à laquelle la commune adhère"]]
df_base = df_base.rename({"Code INSEE de la commune adhérente" : "Code INSEE"}, axis = 1)
```

```python
df_base
```

```python
# Création du df_prix_eau_potable
df_list = []

for annee in range(2012, 2024):
    df1 = pd.read_csv("C:/Users/Antoine/Documents/AquaVision/donnees/brutes/eau_potable/tarifs_AEP_" + str(annee) + ".csv", encoding="Windows-1252", sep=";", low_memory=False)
    
    df1['Id SISPEA de la collectivité'] = df1['Id SISPEA de la collectivité'].str.replace('CollectivityId=', '').astype(int)

    df1 = df1[['Id SISPEA de la collectivité', 'D102.0']]    

    df_temp = pd.merge(df_base, df1, how = 'left', left_on = "Identifiant SISPEA de la collectivité de l'entité de gestion à laquelle la commune adhère", right_on = "Id SISPEA de la collectivité").drop(["Identifiant SISPEA de la collectivité de l'entité de gestion à laquelle la commune adhère", "Id SISPEA de la collectivité"], axis = 1)

    
    df_temp['D102.0'] = df_temp['D102.0'].str.replace(",", ".").astype(float)

    df_temp = df_temp.fillna(0)

    df_temp = df_temp.groupby('Code INSEE').max()

    df_temp['Année'] = annee

    df_temp = df_temp.reset_index()

    df_list.append(df_temp)

# Concatenation de tous les DataFrames accumulés dans la liste
df_prix_eau = pd.concat(df_list, ignore_index = True)

df_prix_eau['D102.0'] = df_prix_eau['D102.0'].apply(lambda x : np.nan if x==0 else round(x , 2))

df_prix_eau
```

```python
# Création du df_prix_assainissement
df_list = []

for annee in range(2012, 2024):
    df1 = pd.read_csv("C:/Users/Antoine/Documents/AquaVision/donnees/brutes/assainissement/tarifs_AC_" + str(annee) + ".csv", encoding="Windows-1252", sep=";", low_memory=False)
    
    df1['Id SISPEA de la collectivité'] = df1['Id SISPEA de la collectivité'].str.replace('CollectivityId=', '').astype(int)

    df1 = df1[['Id SISPEA de la collectivité', 'D204.0']]    

    df_temp = pd.merge(df_base, df1, how = 'left', left_on = "Identifiant SISPEA de la collectivité de l'entité de gestion à laquelle la commune adhère", right_on = "Id SISPEA de la collectivité").drop(["Identifiant SISPEA de la collectivité de l'entité de gestion à laquelle la commune adhère", "Id SISPEA de la collectivité"], axis = 1)

    
    df_temp['D204.0'] = df_temp['D204.0'].str.replace(",", ".").astype(float)

    df_temp = df_temp.fillna(0)

    df_temp = df_temp.groupby('Code INSEE').max()

    df_temp['Année'] = annee

    df_temp = df_temp.reset_index()

    df_list.append(df_temp)

# Concatenation de tous les DataFrames accumulés dans la liste
df_prix_ass = pd.concat(df_list, ignore_index = True)

df_prix_ass['D204.0'] = df_prix_ass['D204.0'].apply(lambda x : np.nan if x==0 else round(x , 2))

df_prix_ass
```

```python
df_prix = pd.merge(df_prix_eau, df_prix_ass, how = 'outer', on = ['Code INSEE', 'Année'])
```
